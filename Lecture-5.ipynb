```python
#practice
def factorial(num):
  fact = 1
  for i in range(1, num+1, 1):
    fact*=i
  return fact

num = int(input('Enter number:  '))
result = factorial(num)
print(f'Result: {result}')
```

```python
def percentage(val, total):
  res = int((val/total)*100)
  print(f'{val} in {total} is: {res}%')

percentage(int(input('Enter value: ')), int(input('Enter Total value: ')))
```

```python
def avg(val, cnt):
  print(f'avg: {val/cnt}')

sum = 0
tCase = int(input('How many number do you want: '))
for i in range(tCase):
  val = int(input('Enter number: '))
  sum+=val
avg(sum, tCase)
```

```python
def validUsername(userName):
  if len(userName) >= 8 and len(userName) <=20:
    for char in userName:
      if char.isalnum():
        continue
      elif char == ' ':
        return 'Alert: Your userName contains white-space.'
      else:
        return 'Alert: Your userName contains special character'
    return 'Username is accepted'
  else:
    return 'Alert: Username should be 8-20 character'


attempt = 0
while True:
  string = input('Enter Username: ')
  output = validUsername(string)
  if output == 'Username is accepted':
    print(output)
    break
  elif attempt > 2:
    print('Maximum attempt reached')
    break
  else:
    print(output)
  attempt+=1
```
