```python
#problem 2
n = int(input("Enter a number: "))
sum = 0
i = 1
while(i<=n):
  sum+=i
  i+=1
print(f'sum of 1 to {n} is: {sum}')
```

```python
#problem 4
n = int(input("Enter a number to get factorial: "))
i = 1
fact = 1
while(i<=n):
  fact*=i
  i+=1
print(f'factorial of {n} is: {fact}')
```

```python
#problem 5
n = 10
i = 1
while(i<=10):
  print(f'{n} x {i} = {n*i}')
  i+=1
```

```python
#problem 6
a = [36, 69,32, 24, 45]
print(f'list: {a}')
n = len(a) - 1
i = 0
while i <= n//2:
  a[i], a[n-i] = a [n-i], a[i]
  i+=1
print(f'reverse list: {a}')
```

```python
#problem 7
n = int(input("Enter a digit: "))
sum = 0
temp = n
while(n>0):
  rem = n % 10
  sum+=rem
  n = int(n/10) #it has an alternative. we can use // insted of type cast to integer n//=10
print(f'sum of digit {temp} = {sum}')
```

```python
#problem 8
a = [36, 69,32, 24, 45]
i = 1
max = a[0]
while i < len(a):
  if max < a[i]:
    max = a[i]
  i+=1
print(f'max value: {max}')
```
