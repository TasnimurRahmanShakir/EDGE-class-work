```python
str_ = 'A quick brown fox jums over the lazy dog'
i = 0
def isVowel(character):
  vowel = 'aeiouAEIOU'
  if vowel.find(character)!= -1:
    return character
  return -1


while i<len(str_):
  if isVowel(str_[i])!=-1:
    print(str_[i])
  i+=1

```

```python
#string[start:end:step]
s = input('Enter your word: ')

revString = lambda s : s[::-1]

if s == revString(s):
  print("Palindrome")
else: print("Not Palindrome")
```

```python
s = input('Enter your string: ')
if s < 'banana':
  print(f'{s} comes before banana')
elif s > 'banana':
  print(f'{s} comes after banana')
else:
  print(f'{s} is equal to banana')
```

```python
s = 'hello World'
print(s.capitalize())
print(s.casefold()) #aggresively convert the string into lower case
print(s.center(12,'*')) #the number should be the greater than string length
print(s.endswith('World'))
print(s.endswith('lo',0,5)) #it checks string[0:5] and check it ends with 'lo' or not
print(s.endswith(('.py','.docx','.text'))) #it checks the string ends with any of them or not
print(s.find('d')) #if value is present then retrurn 'index number' otherwise -1
print(s.lstrip('he')) #remove the occurance on the left of the string
print(s.rstrip('d')) #remove all the occurance on the right of the string
print(s.replace('World', 'New World'))
print(s.strip('hde')) #remove all the character of the string if any of the (hde) present in the string at the start and end
print(s.upper())

#dir(s)
```
