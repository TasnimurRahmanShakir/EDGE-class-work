```python
words = "Candidates, who will take admission by visiting GUB Campus, must submit the attested /photocopies of all Certificates, Testimonials & Transcripts/Mark Sheets, National ID Card/Birth Registration Certificate/Copy of Passport of the candidate, National ID Card of Father/Mother, 2 (two) copies of passport-sized photograph to the GUB Admissions Office at the time of taking admission. Candidates, who will take admission via online, would require sending the scanned copies of above-mentioned documents to the email address: admission@green.edu.bd. Please note that discrepancy in any document of a student may lead to cancellation of admission of the concerned student."

wCounts = dict()
for w in words.split(' '):
  if w in wCounts:
    wCounts[w] += 1
  else:
    wCounts[w] = 1

print(wCounts)
```

```python
dict_ = {'a':4, 'b':1, 'c':6, 'd':2, 'e':3}
newDict = {k:v*5 for (k,v) in dict_.items()}
print(newDict)
newDict2 = {k:v for (k,v) in dict_.items() if 'a' in k}
print(newDict2)
newDict3 = {k:('even' if v%2 ==0 else 'odd') for (k,v) in dict_.items()}
print(newDict3)
newDict4 = {k:v for (k,v) in dict_.items() if v%2==0 if v%3 ==0}
print(newDict4)
```
