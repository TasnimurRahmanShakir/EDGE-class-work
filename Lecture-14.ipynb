```python
import numpy as np
a = np.array([1,2,3])
b = np.array([4,5,6])
print(a+b)
print(a*b)
```

```python
array = np.array([1,2,3,4,5,6])
print(f'Array sum: {np.sum(array)}')
print(f'array Mean: {np.mean(array)}')
print(f'Standard Deviation: {np.std(array)}')
```

```python
a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])
print(np.dot(a, b))
```

```python
a = np.array([[1], [2], [3]])
b = np.array([10, 20, 30])
print(a + b)
```

```python
array = np.array([1,2,3,4,5,6])
normalized_data = (array-np.min(array))/(np.max(array)-np.min(array))
print(f'Normalized array: {normalized_data}')
```

```python
a = np.array([1, 2, 3, 4])
b = a[2:4]
b[0] = 100
print(a)
```

```python
#linear algebra
# 3x + y = 9
# x + 2y = 8
# Ax = b
a = np.array([[3,1],[1,2]])
b = np.array([[9],[8]])
print(np.linalg.solve(a,b))
```
