```python
#practice
plainText = input("Enter your message: ")
key = int(input("Enter your key: "))
cipher = ""
for i in plainText:
  newChar = (ord(i) + key)
  cipher = cipher + chr(newChar)
print(f'Cipher Text: {cipher}')
plainText = ""

for i in cipher:
  plainText += chr(ord(i) - key)
print(f'Message: {plainText}')
```

```python
#playfair
key = input('Enter key: ').replace('j', 'i').lower()


keyBox = [[None for _ in range(5)] for _ in range(5)]
i = 0
char = 'a'

for rIndex in range(5):
    for cIndex in range(5):
        if i < len(key):
            keyBox[rIndex][cIndex] = key[i]
            i += 1
        else:
            while char in key or char == 'j':
                char = chr(ord(char) + 1)
            keyBox[rIndex][cIndex] = char
            char = chr(ord(char) + 1)
for row in keyBox:
    print(row)
```

```python
num = input('Enter a number: ')
track = ""

for i in num:
    if i in track:
        continue


    fre = 0
    for j in num:
        if i == j:
            fre += 1

    print(f'Frequency of {i}: {fre}')
    track += i
```
