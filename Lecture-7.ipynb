```python
students = list()
for i in range(5):
  students.append(input(f'Enter student name {i+1}: ').upper())

for student in students:
  print(student)
```

```python
newList = [23, 45, 67, 66, 92, 34, 69]
def isOdd(n):
  if n%2 == 1:
    return n
  else:
    return False

oddNumbers = list(map(isOdd, newList))
print(f'Using Map: {oddNumbers}')

print(f'Using filter: {list(filter(isOdd,newList))}')
```

```python
numbers = list()
for i in range(10):
  numbers.append(int(input()))

print(f'Output: {list(filter(lambda n: n<20 and n>10, numbers))}')
```

```python
words = list()
for i in range(5):
  words.append(input(f'Enter word {i+1}: '))

print([word for word in words if word.endswith('n')])
```

```python
element = input('Enter your value: ')
index = int(input('Where do you want to insert: '))
words.insert(index, element)
print(words)
```

```python
element = input('Enter element that do you want to remove: ')

def removeElement(elmnt):
  words.remove(elmnt) if elmnt in words else print('Element not found')

removeElement(element)
print(words)
```
