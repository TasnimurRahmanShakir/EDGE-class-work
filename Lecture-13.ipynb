```python
import numpy as np
array2d = np.array([[1,2,3],[4,5,6]])
column_sum = np.sum(array2d, axis=0) #
row_sum = np.sum(array2d, axis=1)
print(f'Column sum: {column_sum}\nRow sum: {row_sum}')
```

```python
array = np.array([1,2,3,4,5,6])
split_array = np.split(array,2)
print(f'Split array: {split_array}')
sum = np.sum(split_array,axis=0)
print(sum)
```

```python
zeros_array = np.zeros((3,3))
print(f'Zeros array: {zeros_array}')
```

```python
reshape_array = array.reshape(3,2)
print(f'Reshape array: {reshape_array}')
```
