```python
import math
flag = 0
i = 2
n1 = int(input('Enter a number: '))
while i <= int(math.sqrt(n1)):
  if n1 % i == 0:
    flag = 1
    break
  i+=1

if flag == 1:
  print('Not prime')
else:
  print('Prime')
```

```python
def grade(n):
  if n > 89:
    return 'A'
  elif n > 79:
    return 'B'
  elif n>69:
    return 'C'
  elif n>59:
    return 'D'
  else:
    return 'F'

n1 = int(input('Enter a number: '))
grade(n1)
```

```python
def isVowel(c):
  if c in 'aeiouAEIOU':
    return 1
  else:
    return 0
s = input('Enter your sentence: ')
cnt = 0
for i in s:
  cnt += isVowel(i)
print(f'Total vowel: {cnt}')
```

```python
balance = 0
while True:
  choice = int(input('1. Check Balance \n 2. Withdraw \n 3. Deposite \n 4. Exit\n Make your choice: '))
  if choice == 1:
    print(f'Your account Balance is: {balance}')
  elif choice == 2:
    withdraw = int(input('Enter your withdraw ammount: '))
    if withdraw > balance:
      print('Insufficient Balance')
    else:
      print(f'{withdraw} Tk withdraw successful.')
      balance -= withdraw
      print(f'Your current account Balance is: {balance}')
  elif choice == 3:
      deposite = int(input('Enter your deposite balance'))
      print(f'{deposite} Tk deposite successful.')
      balance += deposite
      print(f'Your current account Balance is: {balance}')
  else:
    break
```
